```python
#2-1
import cv2 #opencv 모듈
import numpy as np

img = cv2.imread('/content/Fig0309(a)(washed_out_aerial_image).tif') #이미지 불러오기
img = cv2.resize(img, dsize=(0, 0), fx=0.25, fy=0.25) # 이미지 크기 줄이기
cv2_imshow(img)

def gamma(f, gamma=1.0):
  f1=f/255.0 #L=256이라고 가정
  return np.uint8(255.0 * (f1 ** gamma)) # 감마 보정 후 다시 0~255로 변환

gc=np.hstack((gamma(img,1.0),gamma(img,3.0),gamma(img,4.0),gamma(img,5.0))) ## np.hstack : NumPy 배열을 '수평(Horizontal)'으로 쌓아주는 함수

cv2_imshow(gc) #결과출력

# 감마 보정을 실행하였습니다. 비선형 연산이 사용되며. 값이 1.0일때, 원본 영상과 동일, 감마 값이 증가 할 수록 더욱 어두워지며, 탁함이 줄고 명암이 뚜렷해지며, 경계선이 잘 보입니다.
```

```python
#2-2
import cv2 #opencv 모듈
import numpy as np

img = cv2.imread('/content/Fig0310(b)(washed_out_pollen_image).tif') #이미지 불러오기
img = cv2.resize(img, dsize=(0, 0), fx=0.25, fy=0.25) # 이미지 크기 줄이기
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)	# 명암 영상으로 변환하고 출력

def piecewise_linear_transformation(img, L1, L2):
    img = img.astype(np.float32)
    result = np.zeros_like(img)

    result[img <= L1] = 0
    mask = (img > L1) & (img < L2)
    result[mask] = ((img[mask] - L1) * (255.0 / (L2 - L1)))
    result[img >= L2] = 255

    return result.astype(np.uint8)

stretch1 = piecewise_linear_transformation(gray, 70, 150)
stretch2 = piecewise_linear_transformation(gray, 50, 120)
threshold = piecewise_linear_transformation(gray, 110, 111)

images = [gray, stretch1, stretch2, threshold]

plt.figure(figsize=(12, 4))
for i in range(4):
    plt.subplot(1, 4, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.axis('off')
plt.tight_layout()
plt.show()


#구간별 선형 변환을 적용한 것으로, 입력 픽셀 값을 여러 구간으로 나누고, 각 구간에서 선형 함수를 적용하여 변환하였습니다.
```

```python
#2-3
import cv2  #opencv 모듈
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/content/Fig0309(a)(washed_out_aerial_image).tif') #이미지 불러오기
img = cv2.resize(img, dsize=(0, 0), fx=0.25, fy=0.25) # 이미지 크기 줄이기

box = np.ones((3,3), np.float32) / 9
box_result = cv2.filter2D(img, -1, box)#컨볼루션 적용
#박스 필터

gaussian_result = cv2.GaussianBlur(img, (5,5), 0.0)#가우시안 필터

sharp1= np.array([
    [ 0, -1,  0],
    [-1,  4, -1],
    [ 0, -1,  0]
], dtype=np.float32)
sharp1_result = cv2.filter2D(img, -1, sharp1)
#샤프닝 필터1

sharp2= np.array([
    [-1, -1, -1],
    [-1,  8, -1],
    [-1, -1, -1]
], dtype=np.float32)
sharp2_result = cv2.filter2D(img, -1, sharp2)
#샤프닝 필터2

dia1 = np.array([
    [-1, 0, 0],
    [ 0, 0, 0],
    [ 0, 0, 1]
], dtype=np.float32)
dia1_result = cv2.filter2D(img, -1, dia1)
#대각선 방향 edge 필터 1

dia2 = np.array([
    [-1, -1,  0],
    [ 0,  0,  1],
    [ 0,  1,  1]
], dtype=np.float32)
dia2_result = cv2.filter2D(img, -1, dia2) #대각선 방향 edge 필터 2

cv2_imshow(box_result)
cv2_imshow(gaussian_result)
cv2_imshow(sharp1_result)
cv2_imshow(sharp2_result)
cv2_imshow(dia1_result)
cv2_imshow(dia2_result)
# 결과 출력

#2-1 파일을 가져와 분석하였으며 박스와 가우시안 필터는 스무딩 효과가 있고, 박스 필터는 정규 마스크, 가우시안 마스크는 거리에 따라 가중치를 부여합니다. 가우시안 필터가 가중치에 따라 더욱 블러링 된 것을 볼 수 있습니다.
#샤프닝 필터는 명암 대비 강조효과가 있고, 두번째 샤프닝 필터가 첫번재 필터에 비해 더욱 중앙의 선명화가 강조되고, 대비가 극명한 것을 볼 수 있습니다.
#에지 마스크는 일종의 미분 연산자, 아주 좁은 지역에서의 값의 변화 측정합니다. 대각선 방향 에지 필터 1은 왼쪽위에서 오른쪽아래 방향으로 특정 기울기 방향의 변화만 강조되고
#대각선 방향 에지 필터 2은 오른쪽위에서 왼쪽 아래 방향으로 반대쪽 대각선 기울기가 강조됩니다.
```

```python
#2-5
import cv2
import numpy as np
import matplotlib.pyplot as plt

!wget "https://raw.githubusercontent.com/kjcheoi-class/CV/main/finger.jpg" -0 finger.jpg
img = cv2.imread('finger.jpg', cv2.IMREAD_GRAYSCALE)

blur = cv2.GaussianBlur(img, (3,3), 0) #가우시안 블러

_, binary = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU) #Otsu 이진화

kernel = np.ones((3,3), np.uint8)
morph = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)# Morphology 열림 연산으로 잡음 제거

cv2_imshow(img)
cv2_imshow(morph)
#결과 출력

#가우시안블러로 노이즈들 제거하고 Otsu 이진화 morphology 열림 연산으로 잡음 제거
```

```python
#2-7
import cv2
import numpy as np
import matplotlib.pyplot as plt

!wget "https://raw.githubusercontent.com/kjcheoi-class/CV/main/Fig0939(a)(headCT-Vandy).tif" -O headCT-Vandy.tif

img = cv2.imread('headCT-Vandy.tif', cv2.IMREAD_GRAYSCALE)
img = cv2.resize(img, dsize=(0, 0), fx=0.25, fy=0.25) # 이미지 크기 줄이기

# 3x3 소벨 연산자 적용
grad_x=cv2.Sobel(img,cv2.CV_32F,1,0,ksize=3)
grad_y=cv2.Sobel(img,cv2.CV_32F,0,1,ksize=3)

sobel_x=cv2.convertScaleAbs(grad_x)	# 절대값을 취해 양수 영상으로 변환
sobel_y=cv2.convertScaleAbs(grad_y)

edge_strength=cv2.addWeighted(sobel_x,0.5,sobel_y,0.5,0)	# 에지 강도 계산

cv2_imshow(img)
print("\nsobelx\n")
cv2_imshow(sobel_x) #sobelx
print("\nsobely\n")
cv2_imshow(sobel_y) #sobely
print("\nedge_strength\n")
cv2_imshow(edge_strength) #edge_strength

#Laplacian
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = np.uint8(np.clip(np.abs(laplacian), 0, 255))
print("\nLaplacian")
cv2_imshow(laplacian)

#Canny
canny=cv2.Canny(img,50,150)	# Tlow=50, Thigh=150으로 설정
print("\nCanny Edge Detector\n")
cv2_imshow(canny)

# Morphological Gradient
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
morph_grad = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
print("\nMorphological Gradient\n")
cv2_imshow(morph_grad)


#Sobel은 x는 수직, y는 수평 방향성을 강조하고
#Laplacian은 노이즈 민감
#Canny는 정확하고 얇은 엣지
#Morphological Gradient 형태 기반으로 외곽선이 잡힌다.
```

```python
#2-8
import cv2
import numpy as np
import matplotlib.pyplot as plt

!wget "https://raw.githubusercontent.com/kjcheoi-class/CV/main/Fig0938(a)(cygnusloop_Xray_original).tif" -O Xray.tif

img = cv2.imread('Xray.tif', cv2.IMREAD_GRAYSCALE)

# 이진화
t,bin_img=cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

se = np.uint8([
    [0,1,0],
    [1,1,1],
    [0,1,0]
])#구조 요소

#Morphological Smoothing
opened = cv2.morphologyEx(img, cv2.MORPH_OPEN, se)
morph_smooth = cv2.morphologyEx(opened, cv2.MORPH_CLOSE, se)

#Smoothing Filter
smooth = cv2.GaussianBlur(img, (5, 5), 0)

# 결과 출력
cv2_imshow(img)

print("Morphological Smoothing")
cv2_imshow(morph_smooth)

print("Smoothing Filter")
cv2_imshow(smooth)

#Morphological Smoothing은 열림 연산 후 닫힘 연산을 수행하고 노이즈는 좀 제거 된 것 같고 Smoothing Filter는 부드럽게 뭉개진 느낌이다.
```
